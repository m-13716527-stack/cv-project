### WORKING EXAMPLE 1: Basic Text Elements

This example shows how to display titles, headers, text, and styled markdown in Streamlit.

````python
# WORKING EXAMPLE 1: Basic Streamlit text elements

# Write the app code to a file
%%writefile example1_basic_text.py
import streamlit as st

# Create a title (largest text)
st.title('My First Streamlit App')

# Create a header (medium text)
st.header('Details of my first Streamlit App')

# Display plain text
st.text('This is my first streamlit app')

# Display styled markdown (with HTML)
st.markdown('<p style="color:blue">A very cool app with blue text!</p>', unsafe_allow_html=True)

print('\n✅ File created: example1_basic_text.py')
print('To run: streamlit run example1_basic_text.py')
````

**What This Example Shows**:
- `st.title()` - Creates the main title
- `st.header()` - Creates section headers
- `st.text()` - Displays plain text
- `st.markdown()` - Displays styled text with HTML/CSS

**Try It**: Run `!streamlit run example1_basic_text.py` in a terminal to see the app.

### WORKING EXAMPLE 2: Interactive Button

This example demonstrates state management - how Streamlit responds to user clicks.

````python
# WORKING EXAMPLE 2: Interactive button with state

%%writefile example2_button.py
import streamlit as st

# Create title
st.title('My Greeting Streamlit App')

# Create a button (returns True when clicked, False otherwise)
btn = st.button('Say Hello')

# Show different text based on button state
if btn:
    st.text('Hello')  # Shown when button is clicked
else:
    st.text('Goodbye')  # Shown by default

print('\n✅ File created: example2_button.py')
print('To run: streamlit run example2_button.py')
````

**What This Example Shows**:
- `st.button()` - Creates a clickable button, returns True when clicked
- Conditional rendering - Display different content based on user interaction
- State management - Streamlit reruns the entire script on each interaction

**Key Concept**: Streamlit reruns your script from top to bottom every time a user interacts with it.

### WORKING EXAMPLE 3: Form with Multiple Inputs

This example shows different input widgets for collecting user data.

````python
# WORKING EXAMPLE 3: Personal details form

%%writefile example3_form.py
import streamlit as st

# Title and instructions
st.title('My Personal Details Form')
st.subheader('Please enter your personal details')

# Text input
name = st.text_input('Name')

# Number input
age = st.number_input('Age', min_value=0, max_value=120)

# Multi-line text input
address = st.text_area('Address')

# Dropdown selection
gender = st.selectbox('Gender', ('Male', 'Female', 'Other'))

# Checkbox
agree = st.checkbox('I agree to the terms and conditions')

# Submit button
submit = st.button('Submit')

# Display results when submitted
if submit:
    if agree:
        st.write(f'Thank you for submitting your details. Mr/Ms {name}, you are {age} years old. You live at {address} and your gender is {gender}.')
    else:
        st.warning('Please agree to the terms and conditions')

print('\n✅ File created: example3_form.py')
print('To run: streamlit run example3_form.py')
````

**What This Example Shows**:
- `st.text_input()` - Single-line text input
- `st.number_input()` - Numeric input with min/max bounds
- `st.text_area()` - Multi-line text input
- `st.selectbox()` - Dropdown selection
- `st.checkbox()` - Boolean checkbox
- `st.write()` - Smart text display (handles different data types)
- `st.warning()` - Display warning messages

**Form Design Tip**: Always validate user input before processing!

### WORKING EXAMPLE 4: Multi-Page App with Sidebar

This example demonstrates navigation between different pages using a sidebar.

````python
# WORKING EXAMPLE 4: Multi-page app with sidebar navigation

%%writefile example4_multipage.py
import streamlit as st

# Main title
st.title('My App with 2 Pages')

# Sidebar navigation
st.sidebar.title('Sidebar')
page = st.sidebar.selectbox('Select page', ('Home', 'About me'))

# Display different content based on selection
if page == 'Home':
    st.header('Home')
    st.write('This is my home page')
    st.write('Welcome to the app!')

    # Add some sample content
    st.subheader('What you can do here:')
    st.write('- Navigate between pages using the sidebar')
    st.write('- Learn about me on the About page')
    st.write('- Explore Streamlit features')
else:
    st.header('About me')
    st.write('This is my about me page')
    st.write('I am learning to build web apps with Streamlit!')

    st.subheader('My Skills:')
    st.write('- Python programming')
    st.write('- Data analysis')
    st.write('- Web app development')

print('\n✅ File created: example4_multipage.py')
print('To run: streamlit run example4_multipage.py')
````

**What This Example Shows**:
- `st.sidebar` - Creates a persistent sidebar
- `st.sidebar.selectbox()` - Dropdown navigation in sidebar
- Page routing - Simple conditional logic for multi-page apps
- `st.subheader()` - Subsection headers

**App Structure Tip**: For complex apps, consider organizing code into functions or separate files for each page.
