## 📚 Section 2: Reading and Displaying Images - Working Example

````python
# WORKING EXAMPLE 2: Read and display an image
import cv2 as cv
from google.colab.patches import cv2_imshow

# Read the image file
img = cv.imread("outing.jpg")

# Display image properties
print("Image Properties:")
print(f"  Shape: {img.shape}")
print(f"  Height: {img.shape[0]} pixels")
print(f"  Width: {img.shape[1]} pixels")
print(f"  Channels: {img.shape[2]} (BGR format)")
print(f"  Data type: {img.dtype}")

# Display the image in Colab
print("\nDisplaying image:")
cv2_imshow(img)

print("\n✅ Image loaded and displayed successfully!")
````

````python
# WORKING EXAMPLE 2b: Compare different image read modes
import cv2 as cv
from google.colab.patches import cv2_imshow

# Read in color (default)
img_color = cv.imread("outing.jpg", cv.IMREAD_COLOR)

# Read in grayscale
img_gray = cv.imread("outing.jpg", cv.IMREAD_GRAYSCALE)

print("Comparison of Read Modes:")
print(f"  Color shape: {img_color.shape} (Height x Width x Channels)")
print(f"  Grayscale shape: {img_gray.shape} (Height x Width only)")

print("\nGrayscale Image:")
cv2_imshow(img_gray)

print("\n✅ Read modes demonstrated!")
````

## 📚 Section 3: Video Playback - Working Example

````python
# WORKING EXAMPLE 3: Play video in Google Colab
from IPython.display import HTML
from base64 import b64encode

# Note: cv.imshow() doesn't work in Colab for video
# We use HTML5 video player instead

# Read video file and encode for browser playback
mp4 = open('video.mp4', 'rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()

print("Video Information:")
print(f"  File size: {len(mp4) / 1024:.1f} KB")
print("\n📝 Note: In Colab, we display video using HTML5 player")
print("   In local Python, you would use cv.imshow() in a loop")

# Display video with HTML5 player
print("\nVideo Player:")
HTML("""
<video width=400 controls>
      <source src="%s" type="video/mp4">
</video>
""" % data_url)
````

````python
# WORKING EXAMPLE 3b: Extract video properties using VideoCapture
import cv2 as cv

# Open video file
video = cv.VideoCapture("video.mp4")

# Get video properties
fps = video.get(cv.CAP_PROP_FPS)
frame_count = int(video.get(cv.CAP_PROP_FRAME_COUNT))
width = int(video.get(cv.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv.CAP_PROP_FRAME_HEIGHT))
duration = frame_count / fps

print("Video Properties:")
print(f"  Resolution: {width}x{height} pixels")
print(f"  Frame rate: {fps:.1f} FPS")
print(f"  Total frames: {frame_count}")
print(f"  Duration: {duration:.2f} seconds")

# Read first frame
ret, first_frame = video.read()
if ret:
    print(f"\nFirst frame shape: {first_frame.shape}")

# Always release the video capture object
video.release()

print("\n✅ Video properties extracted!")
````

## 📚 Section 4: Webcam Access in Colab - Working Example

````python
# ⚠️ Colab-only cell: take_photo() uses a JavaScript browser shim.
# In VS Code / local Jupyter / any non-Colab environment, this cell will
# hang forever (you'll see "running 28.0s" with no output). If that happens,
# you're not in Colab — skip this cell and run `python webcam.py` locally.

# WORKING EXAMPLE 4: Webcam capture in Google Colab
# Note: This uses JavaScript to access browser webcam

from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):
    """Capture a photo from webcam in Colab.
    
    Args:
        filename: Name of the output file
        quality: JPEG quality (0.0 to 1.0)
    
    Returns:
        filename of the saved photo
    """
    js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
    display(js)
    data = eval_js('takePhoto({})'.format(quality))
    binary = b64decode(data.split(',')[1])
    with open(filename, 'wb') as f:
        f.write(binary)
    return filename

print("📷 Webcam Function Ready!")
print("\n📝 How it works:")
print("  1. JavaScript accesses browser webcam")
print("  2. Shows video preview with 'Capture' button")
print("  3. Captures frame and saves as JPEG")
print("\n⚠️ Note: This requires browser webcam permission")
print("\n✅ Webcam function defined!")
````

````python
# ⚠️ Colab-only cell: depends on take_photo() from the previous cell.
# Will hang forever outside Colab. Skip if you're running locally — use webcam.py instead.

# WORKING EXAMPLE 4b: Capture and display webcam photo
from IPython.display import Image

print("Click 'Capture' button when ready...")
print("(Grant webcam permission if prompted)\n")

try:
    filename = take_photo()
    print(f'\n✅ Photo saved to: {filename}')
    
    # Display the captured image
    display(Image(filename))
except Exception as err:
    print(f"\n⚠️ Webcam error: {str(err)}")
    print("This is normal if:")
    print("  - You denied webcam permission")
    print("  - No webcam is connected")
    print("  - You're running on a server without display")
````
